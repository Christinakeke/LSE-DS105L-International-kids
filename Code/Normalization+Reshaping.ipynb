```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import pandas as pd

monthly_average_CNY=pd.read_excel('monthly_average_CNY.xlsx')

# Create a StandardScaler object
scaler = StandardScaler()

# Extract the exchange rate column from the dataframe
exchange_rate = monthly_average_CNY['CNY/USD']

# Reshape the exchange rate data to a 2D array
exchange_rate_2d = exchange_rate.values.reshape(-1, 1)

# Fit the scaler to the exchange rate data
scaler.fit(exchange_rate_2d)

# Normalize the exchange rate data
exchange_rate_normalized = scaler.transform(exchange_rate_2d)

# Create a new dataframe with the same columns as the original dataframe
normalized_dataframe_CNY = pd.DataFrame({
    'Date': monthly_average_CNY['index'],
    'Exchange Rate': exchange_rate_normalized.flatten()
})

# Print the new dataframe
print(normalized_dataframe)
```
